```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reddit_dad_jokes = pd.read_csv('/kaggle/input/reddit-dad-jokes/reddit_dadjokes.csv')
rjokes_reddit = pd.read_csv('/kaggle/input/selflearning/train.tsv', delimiter='\t', error_bad_lines=False)
```

```python
rjokes_reddit.columns = ['score', 'joke']
```

```python
def get_joke_word_count(list_of_jokes):
    word_count = []
    for joke in list_of_jokes:
        word_count.append(len(joke.split()))
        
    return word_count

def plot_word_count_distribution(frame):
    plt.hist(frame['wc'], bins=1000, color='blue', edgecolor='black')

    plt.xlabel('word_count')
    plt.ylabel('Frequency')
    plt.title('Histogram of Word Count')

    plt.show()
    
    return
```

```python
rjokes_reddit_wc = get_joke_word_count(rjokes_reddit['joke'].tolist())
dad_jokes_reddit_wc = get_joke_word_count(reddit_dad_jokes['joke'].tolist())
rjokes_reddit['wc'] = rjokes_reddit_wc
reddit_dad_jokes['wc'] = dad_jokes_reddit_wc
reddit_dad_jokes = reddit_dad_jokes[reddit_dad_jokes['wc'] < 35]
rjokes_reddit = rjokes_reddit[rjokes_reddit['wc'] < 35]
```

```python
plot_word_count_distribution(reddit_dad_jokes)
```

```python
plot_word_count_distribution(rjokes_reddit)
```
